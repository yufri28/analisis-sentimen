```python
import pandas as pd
import re

# Load data dari file CSV
data = pd.read_csv('D:\Semester VIII\program_skripsi\dataset.csv')

# Fungsi untuk mencari kata tidak standar
# def cari_kata_tidak_baku(stopword_removal):
#     # Load kamus baku
#     with open('indonesian-words.txt', 'r') as file:
#         kamus_baku = file.read().splitlines()
#     # Pisahkan setiap kata pada stopword_removal
#     kata = re.findall(r'\b\w+\b', stopword_removal)
#     # Cari kata-kata tidak baku
#     kata_tidak_baku = [k for k in kata if k.lower() not in kamus_baku]
#     # Kembalikan daftar kata tidak baku
#     return kata_tidak_baku

def cari_kata_tidak_baku(stopword_removal):
    # Load kamus baku dari KBBI.csv
    kamus = pd.read_csv('kbbi.csv', header=None, names=['kata'])
    kamus_baku = set(kamus['kata'].str.lower())
    # Pisahkan setiap kata pada teks
    kata = re.findall(r'\b\w+\b', stopword_removal)
    # Cari kata-kata tidak baku
    kata_tidak_baku = [k for k in kata if k.lower() not in kamus_baku]
    # Kembalikan daftar kata tidak baku
    return kata_tidak_baku


# Buat daftar kata tidak standar dari dataframe stopword_removal
list_kata_tidak_baku = []
for i, row in data.iterrows():
    stopword_removal = row['stopword_removal']
    kata_tidak_baku = cari_kata_tidak_baku(stopword_removal)
    list_kata_tidak_baku.extend(kata_tidak_baku)

# Hapus duplikat kata tidak standar
list_kata_tidak_baku = list(set(list_kata_tidak_baku))

# Simpan list kata tidak standar ke dalam file xlsx
df = pd.DataFrame({'Kata Tidak Baku': list_kata_tidak_baku})
df.to_excel('list_kata_tidak_baku.xlsx', index=False)
```
